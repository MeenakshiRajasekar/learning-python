# Set Functions

```python
# to create empty set function
ex1 = set()
type(ex1)

#to create set ds
ex2 = {1,2,3,'meena', 'AI'}
print(ex2)
```

```python
input1 = {1,2,3,'meena', 'abi'}
input2 = {'Meena', 'meena', 'aarthi'}
tmp = {-000, -111}
cache_1 = set()
print('\n####### add function to add the new elements to set #######')
input1.add('megala')
print(input1)
print('\n####### update function add multiple elements to set #######')
input1.update({'dheerthi', 'raji'})
print(input1)

print('\n####### remove function remove  elements from set - throw  error if element is not there #######')
input1.remove('megala')
print(input1)

print('\n####### discard function remove  elements from set it wont throw any error if element is not there #######')

input1.discard('mk')
print(input1)

print('\n####### pop function to remove random element from set #######')
input1.pop()
print(input1)

print('\n####### clear function to clear all element from set #######')
print(f'Before using clear function : {tmp}')
tmp.clear()
print(f'After using clear function : {tmp}')

print('\n####### union function to return the  combined element in both set (wont repeat the same element) #######')
uni = input1.union(input2)
print(f"combined elements in input1 and input2 are: {uni}")

print('\n####### intersection function to return the  common element in both set (case sensitive) #######')
intersect = input1.intersection(input2)
print(f"Common elements in input1 and input2 are: {intersect}")

print('\n####### difference function to return the  element present only in first set  #######')
diff = input2.difference(input1)
print(f"elements present only in input2 are: {diff}")

print('\n####### symmetric_difference function to return the element present in either set not both(neglects common elements) #######')
sym_diff = input2.symmetric_difference(input1)
print(f"elements present only in either set but not both sets are: {sym_diff}")

print('\n####### intersection_update function to update set with common element #######')
print(f'element present in input2 before updating with intersection_update are : {input2}')
input2.intersection_update(input1)
print(f"input2 set is updated with commont elements: {input2}")

input2. update(['raji', 'sakthi', 2,4,5,6])

print('\n####### symmetric_difference_update function to update set with common element #######')
print(f' input 1 set: {input1}')
print(f' input 2 set: {input2}')
input1.symmetric_difference_update(input2)
print(f" input1 set is updated with elements present only in input1 and input2 not in common: {input1}")


print('\n####### subset function to check whether all the elements present in another set #######')
print(f'input2 is a subset of input1 : {input2.issubset(input1)}')

print('\n####### superset function to check whether set contains  all the element present in another set #######')
print(f'input1 is a super set of input2 : {input1.issuperset(input2)}')

print("\n####### copy function #######")
cache_1 = input1.copy()
print(cache_1)

      
```

## set functions


```python
ip1 = {'Python', 'Java', 'SQL', 'Spring', 'Microservices', 'Docker', 'Kubernetes'}
ip2 = {'Python', 'Java'}
ip3 = {'AWS', 'GCP', 'Azure'}

print('\n####### subset function to check whether all the elements present in another set #######')
print(f'input2 is a subset of input1 : {ip2.issubset(ip1)}')

print('\n####### superset function to check whether set contains  all the element present in another set #######')
print(f'input1 is a super set of input2 : {ip1.issuperset(ip2)}')

print('\n####### disjoint function to check whether there is no common elements between two sets #######')
print(f'input1 and input2 are disjoint sets : {ip1.isdisjoint(ip2)}')
print(f'input2 and input3 are disjoint sets : {ip2.isdisjoint(ip3)}')
```
