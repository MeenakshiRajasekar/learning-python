# Hangman Game

```python
import random

def play_hangman():
    
    words = ['pen', 'paper', 'note', 'ruler', 'eraser']
    selected = random.choice(words)
    hint = ['_'] * len(selected)
    guessed_letter = []
    
    lives = 6
    
    while '_' in hint and lives > 0:
        
        print(f" hint is : {' '.join(hint)}")
        print(f" lives : {lives}")
        print(f" guessed letters : {', '.join(guessed_letter)}")
    
        guess = input("enter your guess (single character between a-z) : ").lower()
    
        if not guess.isalpha() or len(guess) != 1:
            print("Invalid input")
            continue

        if guess in guessed_letter:
            print("your already guessed this letter")
            continue
        
        guessed_letter.append(guess)
            
    
        if guess in selected:
            for i in range(len(selected)):
                if(selected[i] == guess):
                    hint[i] = guess
        else:
            print("your guess letter is not in chosen word")
            lives -= 1
        print('**********************************')
    
    if  '_' not in hint:
        print(f"Congrats! you won the game, the word is :  {selected.upper()}")
    else:
        print(f"you lose the game, the word is: {selected.upper()}")
                
    
```

```python
play_hangman()
```
