## while loop example 1

```python
import random

secret_number = random.randint(1,10)
while True:
    input_number = int(input("enter a number within 10 :"))
    if(input_number == secret_number):
        break
print("entered input matches the secret number")
```

## while loop example 2

```python
print("creating a strong password requires atleast 5 characters")

while True:
    user_input = input("New password :")
    if(len(user_input)>=5):
        break
    else:
        print("Input requires 5 character atleast")
print("Password created successfully")
```

## for loop example 1

```python
# find the factors of number

result = []
input_number = int(input("Enter the number: "))
for i in range(1,input_number+1):
    if(input_number % i ==0):
        result.append(i)
print(result)
```

## if else example 1

```python
# printing odd and even numbers 

input_number = int(input("Enter the range: "))
                   
for i in range(1, input_number):
    if(i%2==0):
        print(f"{i} is even number")
    else:
        print(f"{i} is odd number")
```

## if elif else example 1

```python
ip = int(input ("Enter a number : "))

if(ip > 0):
    print("Given number is positive")
elif(ip < 0):
    print("Given number is negative")
else:
    print("Given number is zero")
```

## if elif else example 2

```python
a = int(input ("Enter first number : "))
b = int(input ("Enter second number : "))
if(a > b):
    print("a is greater than b")
elif(a < b):
    print("a is lesser than b")
else:
    print("a and b are equal")
```

## user defined functions with for loop

```python
#find a number in the given tuple

def find_no(x):
    input_tuple = [1, 2, 3, 4, 5]
    for i in input_tuple:
        if (i ==x):
            print ("found the number")
            return
            

    print("number not found")
        
```

```python
find_no(2)
```

```python
#factorials of a number

def find_factorials(n):
    factorial = 1
    for i in range(1, n+1):
        factorial = factorial * i
        
    print(factorial)
```

```python
find_factorials(5)
```
