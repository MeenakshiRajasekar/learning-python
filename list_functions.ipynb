## Python data structure list inbuilt functions:

```python
input1 = ['a', 'b', 'c']
input2 = [1, 8, 3, 4,0, 2,5]

print('#####  append fucntion adds one element to list ####')

input1.append(44)
print(input1)

print('#### extend function can add multiple values to the list ####')

input1.extend([1,2])
print(input1)

input1.extend([1])
print(input1)

print('\n#### insert function  insert one element in the given position to the list ####')

input1.insert(0, 'main')
print(input1)

print('\n#### remove function  remvoves first occurred element in the  list ####')

input1.remove(1)
print(input1)

print('\n#### pop function  remvoves element in the given position of list ####')

input1.pop(4)
print(input1)

input1.append(1)
print('\n#### index function  prints the index of  element in the list ####') # first occurrence
print(input1)
ind = input1.index(1)
print(f'Index of element 1 in the list : {ind}')

print('\n#### count function  prints the count of  element in the  list ####')
print(input1)
cnt = input1.count(1)
print(f'Count of element 1 in the list : {cnt}')

print('\n#### sort function  sorts the element in the  list ####') # sorting not allowed in heterogeneous list
print(f'Before sorting: {input2}')
input2.sort()
print(f'After sorting: {input2}')

print('\n#### reverse function  reverse the element in the  list ####') 
print(f'Before reverse: {input1}')
input1.reverse()
print(f'After reverse: {input1}')

print('\n#### copy function  copies the element in the given list and creates new list ####')

input3 = input1.copy()
print(f'Input 1 has been copied and created the list as: {input3}')
```
