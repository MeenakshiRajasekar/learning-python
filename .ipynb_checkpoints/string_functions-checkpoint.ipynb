# String functions

```python
print("Let's play with string functions")
user_input = input("Enter your input:")
input2 = input("Enter one more input")
format_input = input("Enter your input for formatting")
format_map_input = input("enter format map input")
map_value = {"name": "Meenakshi", "age": 27}
capitalize = user_input.capitalize()
casefold = user_input.casefold()
centered = user_input.center(100)
count = user_input.count("a")
encode = user_input.encode()
endWith = user_input.endswith("n")
tabspace = input2.expandtabs(2)
findString = input2.find("Me")
fmt_ouput = format_input.format("Meena", 27)
map_fmt = format_map_input.format_map(map_value)
print(len(user_input))
print(capitals)
print(casefold)
print(centered)
print (count)
print(encode)
print (endWith)
print(tabspace)
print(findString)
print(fmt_ouput)
print(map_fmt)
```
