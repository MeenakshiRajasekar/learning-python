## Dictionary functions

```python
## dictionary data structure stores data in 2d format which is key-value pair. Keys act as unique identifiers and stores values.
## to create empty dictionary
dictds = {}


dict1 = {'name' : "Meenakshi", "age" : 21}
type(dict1)

dict2 = {'name' : 'Meenakshi', 'marks' : [90, 92, 85, 97, 88]}
```

```python
print("##### dict function to create a new dictionary ds #####")
student_data = dict(name= 'Meenakshi', age = 27, total = 490, average = 98.2 )
print(student_data)

print("\n##### copy function to copy the dictionary #####")
dictcp = {}
print(dictcp)
dictcp = dict1.copy()
print(f"Dictionary values after using clear function: {dictcp}")

print("\n##### clear function to clear the dictionary #####")
print(f"Dictionary values before using clear function: {dict1}")
#dict1.clear()
print(f"Dictionary values after using clear function: {dict1}")

print("\n##### fromKeys function used to create a dictionary with the specified values #####")
keys = ['m1', 'm2', 'm3']
dictnew = dict.fromkeys(keys, 100)
print(dictnew)

print("\n##### get function used to fetch the values of specified key #####")
print(f"value stored in name key is : {dict1.get('name')}")


print("\n##### item function used to show the key-value pair #####")
print(dict1.items())
print(student_data.items())

print("\n##### key function used to show the key set of the dictionary #####")
print(f'key set of dict1 : {dict1.keys()}')
print(f'key set of student_data : {student_data.keys()}')

print("\n##### pop function used to remove  the specified key value pair #####")
dict2.pop('name')
print(f"after using pop fucntion :  {dict2}") 

print("\n##### pop item function used to remove  the last inserted key-value pair #####")
print(f"before using pop item in dict1 : {dict1}")
dict1.popitem()
print(f"after using pop item in dict1 : {dict1}") ## -> last inserted age key pair removed

print("\n##### setdefault  function used to retrieve  the value of key otherwise return the mentioned default value #####")
print(student_data.setdefault('name' , "xyx"))

print("\n##### update  function used to update  the dictionary with key value #####")
print(f'Before updating the student_data :  {student_data}')
student_data.update ({'course' : "EEE"})
print(f'After updating the student_data :  {student_data}')

print("\n##### value   function used to fetch values present in dictionary #####")
student_data.values()
```
