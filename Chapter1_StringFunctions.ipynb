## String Functions

### Case conversions

```python
input1 = 'Hello'
input2 = 'Hi Meenakshi'

# to upper case
print('\n###### to upper case ######')
print(input1.upper())
print(input2.upper())

print('\n###### to lower case ######')

print(input1.lower())
print(input2.lower())

print('\n###### convert as title format - first letter of all words in input string ######')

print(input1.title())
print(input2.title())

print('\n###### do capitalize - first letter of the given input string ######')

print(input1.capitalize())
print(input2.capitalize())

print('\n###### casefold - more advanced lower case ######')

print(input1.casefold())
print(input2.casefold())
```

### Removing space

```python
input1 = '              Hi, Meenakshi Rajasekar                     '
input2 = '              Proitbridge AI course                       '


print('\n###### to remove the space in common ######')

print(input1.strip())
print(input2.strip())

print('\n###### to remove the space in right side ######')

print(input1.rstrip())
print(input2.rstrip())

print('\n###### to remove the space in left side ######')

print(input1.lstrip())
print(input2.lstrip())
```

### Find and Index functions

```python
## find wont throw any error if the string is not present but index function will throw error otherwise returns the index no

input1 = 'Hello Meenakshi Hello'
input2 = 'course string functions chapter course'


print('\n###### to find the given string in input ######')
print(input1.find('Hello'))
print(input2.find('chapt00'))

print('\n###### to return the last occurrence of string ######')
print(input1.rfind('Hello'))
print(input2.rfind('chapt'))

print('\n###### to return index similar to find function ######')
print(input1.index('Hello'))
print(input2.index('chapt'))

print('\n###### to return the number of occurrence of string ######')
print(input1.count('Hello'))
print(input2.count('course'))

print('\n###### to check whether  the  input starts with given string ######')
print(input1.startswith('Hello'))
print(input2.startswith('cpe'))

print('\n###### to check whether  the  input ends with given string ######')
print(input1.endswith('lo'))
print(input2.endswith('init'))

print('\n###### to replace   the  input  with given string ######')
print(input1.replace('H','Z'))
print(input2.replace('course', 'class'))

print('\n###### to fill   the  input  with zeros ######')
print(input1.zfill(50))
print(input2.zfill(50))

print('\n###### to return last index similar to rfind function ######')
print(input1.rindex('Hello'))
print(input2.rindex('cou11'))
```

### Split and join functions

```python

input1 = 'apple banana carrot'
input2 = "home/user/documents/report.pdf"
input3 = ["apple", "banana", "carrot"]

print('\n###### to split the given input to list ######')
print(input1.split())
print(input2.split('/',1))

print('\n###### to split the given input to list from right side ######')
print(input1.rsplit())
print(input2.rsplit('/',1))

print('\n###### to join the given list to string ######')
print('-'.join(input1))
print('-'.join(input3))

print('\n###### to partition the given string ######')
print(input2.partition("/"))
print(input1.partition(' '))

print('\n###### to partition the given string from right side ######')
print(input2.rpartition("/"))
print(input1.rpartition(' '))

print('\n###### to split the given input in new lines ######')
print('Hi\n hello'.splitlines())
print('Hi\nMeenakshi'.splitlines())
```

### Validating the input 

```python
input1 = 'Meenakshi'
input2 = 'meenakshi rajasekar !'
input3 = '123456'
input4 = 'MEENA'
input5 = 'Hello World'

print('\n###### to check whether the given input  is alphanumeric ######')
print(input1.isalnum())
print(input3.isalnum())

print('\n###### to check whether the given input  is only alphabetic ######')
print(input1.isalpha())
print(input2.isalpha())

print('\n###### to check whether the given input  is only digits ######')
print(input1.isdigit())
print(input3.isdigit())

print('\n###### to check whether the given input  is only lower charactera ######')
print(input1.islower())
print(input2.islower())

print('\n###### to check whether the given input  is only upper characters ######')
print(input1.isupper())
print(input4.isupper())

print('\n###### to check whether the given input  is having space ######')
print(input1.isspace())
print(input2.isspace())

print('\n###### to check whether the given input  is in title format ######')
print(input2.istitle())
print(input5.istitle())


print('\n###### to check whether the given input  is in numerics ######')
print(input1.isnumeric())
print(input3.isnumeric())

print('\n###### to check whether the given input  is in decimals ######')
print(input1.isdecimal())
print(input3.isdecimal())

print('\n###### to check whether the given input  can be identifier ######')
print('input1'.isidentifier())
print('###'.isidentifier())

print('\n###### to check whether the given input  can be printable ######')
print('input1'.isprintable())
print('//\n'.isprintable())
```
