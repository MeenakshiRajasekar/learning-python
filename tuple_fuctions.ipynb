## Tuple Functions

```python
# to create empty tupleds:
tupleds = ()
type(tupleds)

# initialize with values

tupleds = (1, 2, 'Meenakshi')
## tupleds[0] = ['Meenakshi'] => throws error tuple is immutable 

## tuple is 1d data structure, heterogeneous allows indexing and slicing
## tuple object is immutable

print(tupleds[1:3])
```

```python
input1 = (1,2,3,55,99,1,2,0,0,1,2,99)

print("##### tuple count function returns the number of occurrence of element #####")
print(input1.count(1))

print("##### tuple index function returns the index number of first occurrence of element #####")
print(f"element 1 index is : {input1.index(1)}")
print(f"element 1 index is : {input1.index(55)}")
```

###### MRO - Method Resolution Order is the order in which Python searches for a method or attribute in a class hierarchy, especially when inheritance  (multiple inheritance) is used.
