## Dataframes

```python
import pandas as pd

dict1 = {"Name" : ["Meenakshi", "Nakshathra", "Shivani"],
         "Age"  : [27, 28, 29],
         "Place" : ["Chennai", "Hyderabad", "Bangalore"]
        }

dict2 = {"EmployeeName" : ["Arthi", "Arun", "Anu"],
         "Tech-stack" : ["Java backend", "React Frontend", "AWS cloud"],
         "Experience" : [4, 5, 6]
        }

dict3 = {"Products" : ["Laptop", "Mobile", "Keyboard"],
         "Quantity" : [1, 4, 1],
         "Price" : [100000, 200000, 1900]
        }

DF1 = pd.DataFrame(dict1)
DF2 = pd.DataFrame(dict2)
DF3 = pd.DataFrame(dict3)

print(f"\nStudent Table : \n{DF1.to_string(index=False)}")
print(f"\nEmployee Table : \n{DF2.to_string(index=False)}")
print(f"\nInventory Table : \n{DF3.to_string(index=False)}")
```
